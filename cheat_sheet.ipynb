Aby model oparty na regresji logistycznej mógł pełnić rolę klasyfikatora wieloklasowego należy zastosować regresję softmax 
dodając do modelu parametr 'multi_class = "multinominal"'

```python
from sklearn.linear_model import LogisticRegression

softmax_reg = LogisticRegression(multi_class="multinomial")
```

Zz regularyzację modelu odpowiada parametr C -> im większa jego wartość tym mniejszy stopień regularyzacji modelu<br>
Parametr ten musi być większy od 0<br>
Wartość domyślna to C=1

#### Rodzaje regularyzacji
- Regresja grzbietowa -> dla danych przeskalowanych gdzie wartości wag dla wszystkich cech rozkładając się równomiernie
- Regresja LASSO -> dąży do eliminacji wag dla najmniej istotnych cech poprzez utworzenie modelu rzadkiego
- Elastycznej siatki -> parametr r = 1 -> regresja LASSO a r = 0 -> grzbietowa - pozwala ona na wyśrodkowanie regresji pomiędzy grzbietową a LASSO

Solver a rodzaj regresji i kary:<br>
- ‘lbfgs’ - [‘l2’, None]
- ‘liblinear’ - [‘l1’, ‘l2’]
- ‘newton-cg’ - [‘l2’, None]
- ‘newton-cholesky’ - [‘l2’, None]
- ‘sag’ - [‘l2’, None]
- ‘saga’ - [‘elasticnet’, ‘l1’, ‘l2’, None]
